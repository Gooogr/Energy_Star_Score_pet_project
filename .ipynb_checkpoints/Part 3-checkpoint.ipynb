```python
import pandas as pd
pd.set_option('display.max_columns', 60)
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import seaborn as sns

from IPython.core.pylabtools import figsize 
from IPython.display import Image

from sklearn.externals import joblib
from sklearn.preprocessing import Imputer, StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import tree

import lime
import lime.lime_tabular
```

## Загрузка результатов предыдущего шага

Оптимизированная модель градиентного бустинга:

```python
model = joblib.load('model/FINAL_model.pkl')
```

Метки и признаки:

```python
train_features = pd.read_csv('data/training_features.csv')
test_features = pd.read_csv('data/testing_features.csv')
train_labels = pd.read_csv('data/training_labels.csv')
test_labels = pd.read_csv('data/testing_labels.csv')
```

```python
imputer = Imputer(strategy = 'median')
imputer.fit(train_features)

X = imputer.transform(train_features)
X_test = imputer.transform(test_features)

scaler = StandardScaler()
scaler.fit(X)

X = scaler.transform(X)
X_test = scaler.transform(X_test)

y = np.array(train_labels).reshape(-1) #get 1-D array
y_test = np.array(test_labels).reshape(-1)
```

Оценка ошибки модели по МАЕ:

```python
def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))
```

```python
y_pred = model.predict(X_test)
print('MAE: {}'.format(mae(y_test, y_pred)))
```

## Признаки и их влияние на модель

**Рассмотрим датафрейм со всеми признаками, которые были использованы в модели.**

```python
features_importance_frame = pd.DataFrame({'Feature name': train_features.columns, 
                                          'Value': model.feature_importances_}).sort_values('Value', 
                                                                                            ascending=False)
features_importance_frame.set_index('Feature name', inplace = True)
```

```python
features_importance_frame.head()
```

```python
figsize(12, 7)
features_importance_frame.head(10).plot(kind = 'barh')
plt.ylabel(' ')
plt.xlabel('Важность (корреляция) признака', size = 14)
plt.yticks(size = 12)
plt.title('Top 10 ключевых признаков модели', size = 16);
```

**Оценим влияние количества отбираемых признаков на результат**

```python
model_results = []
for num in range(1, len(features_importance_frame)):
    # Отбор необходимых признаков из исходного датафрейма
    important_features_names = features_importance_frame.head(num).index
    train_features_short = train_features.loc[:, important_features_names]
    test_features_short = test_features.loc[:, important_features_names]

    X_short = imputer.fit_transform(train_features_short)
    X_test_short = imputer.fit_transform(test_features_short)
    
    # подготовка новой модели, с использованием гиперпараметров исходной
    optimized_params = model.get_params()
    shorted_model = GradientBoostingRegressor()
    shorted_model.set_params(**optimized_params)
    shorted_model.fit(X_short, y)

    y_pred_short = shorted_model.predict(X_test_short)
    
    mae_coef = mae(y_test, y_pred_short)
    model_results.append(mae_coef)
```

```python
figsize(10, 6)
plt.plot(range(1, len(features_importance_frame)), model_results)
plt.xlabel('Число признаков в выборке', size = 14)
plt.ylabel('Значение MAE', size = 14)
plt.xticks(size = 12)
plt.yticks(size = 12)
plt.title('Влияние числа признаков на качество предсказания', size = 16);
```

```python
residuals = abs(y_pred_short - y_test)

# Exact the worst and best prediction
wrong = X_test_short[np.argmax(residuals), :] # берем строку с лучшим\худшим пресдказанием
right = X_test_short[np.argmin(residuals), :]
```

```python
explainer = lime.lime_tabular.LimeTabularExplainer(training_data = X_short,
                                                   mode = 'regression',
                                                   training_labels = 'y',
                                                   feature_names = list(important_features_names)
                                                  )
```

```python
print(shorted_model.predict(wrong.reshape(1, -1)))
print(y_test[np.argmax(residuals)])

wrong_exp = explainer.explain_instance(data_row = wrong,
                                       predict_fn = shorted_model.predict,
                                      num_features = 10)
```

```python
wrong_exp.as_pyplot_figure();

# wrong_exp.show_in_notebook(show_predicted_value=False);
```

```python
print(shorted_model.predict(right.reshape(1, -1)))
print(y_test[np.argmin(residuals)])

right_exp = explainer.explain_instance(data_row = right,
                                       predict_fn = shorted_model.predict,
                                      num_features = 10)
```

```python
right_exp.as_pyplot_figure();
```
